```python
import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16 , "load_in_8bit":True})

```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.embeddings import LangchainEmbedding

embed_model = LangchainEmbedding(
  HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
)
service_context = ServiceContext.from_defaults(
    chunk_size=200,
    llm=llm,
    embed_model=embed_model
)
```

```python
import warnings
warnings.filterwarnings("ignore")
import os
import pandas as pd
from googleapiclient.discovery import build
from google.colab import auth
from google.colab import files
import re
import ast
import os
from bs4 import BeautifulSoup
from concurrent.futures import ThreadPoolExecutor
```

```python
def process_file(directory, filename):
    """
    Function to process each file.
    Extracts text from HTML content and writes it back to the file.
    """
    if filename.endswith(".txt"):
        file_path = os.path.join(directory, filename)

        with open(file_path, 'r', encoding='utf-8') as file:
            html_content = file.read()

        # Try with 'lxml' parser first
        try:
            soup = BeautifulSoup(html_content, 'lxml')
        except Exception as e:
            print(f"Error with lxml parser for {filename}: {e}. Trying with 'html5lib'.")

            # If lxml fails, try with 'html5lib'
            try:
                soup = BeautifulSoup(html_content, 'html5lib')
            except Exception as e:
                print(f"Both parsers failed for {filename}: {e}. Skipping file.")
                return

        text = soup.get_text()

        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(text)

        print(f'Converted {filename} to plain text format.')


def extract_information(content):
      with open('/content/data/temp.txt', 'w', encoding='utf-8', errors='ignore') as file:
          file.write(content)
      documents = SimpleDirectoryReader("/content/data/").load_data()
      index = VectorStoreIndex.from_documents(documents, service_context=service_context)
      query_engine = index.as_query_engine()
      response = query_engine.query("""Generate a Python dictionary output structured as follows:

'Name': Gather all names of mentioned individuals from the text.
'Age': Include the respective ages for each individual; for missing age data, use 'No data found'.
'Position Held': Present positions linked to each individual, separated by commas \
 if multiple positions are held. Ensure the dictionary strictly adheres to this \
 structure, assembling information directly extracted from the text.
      """)
      return response.response

def extract_company_name(content):
      text = content
      match = re.search(r'COMPANY CONFORMED NAME:\s+([\w\s&]+)', text)
      if match:
          company_name = match.group(1)
          company_name = re.sub(r'\n\t\tCENTRAL INDEX KEY.*', '', company_name)
          return company_name
      else:
          return None
def extract_matched_texts(content):
    pattern = re.compile(r'(item\s*10[.:])(.*?)(item\s*11[.:])', re.DOTALL | re.IGNORECASE)
    matches = pattern.finditer(content)

    extracted_texts = []
    for match in matches:
        extracted_text = match.group(2).strip()
        combined_text = ' '.join(extracted_text.split())
        extracted_texts.append(combined_text)
    if len(extracted_texts) > 1:
        extracted_texts.pop(0)
    if extracted_texts:
        return extracted_texts
    else:
        return ["No matches found"]



# Authenticate
auth.authenticate_user()
# Create Google Drive service
drive_service = build('drive', 'v3')
folder_id = ''
df = pd.DataFrame(columns=['File Name', 'Company Name', 'Year', 'Quarter', 'Executive Details'])
page_token = None
count = 0
while True:
    # List files in the folder with a page token
    results = drive_service.files().list(q=f"'{folder_id}' in parents and trashed=false",
                                         fields="nextPageToken, files(id, name)",
                                         pageSize=100,
                                         pageToken=page_token).execute()
    items = results.get('files', [])

    for item in items:
        print(count)
        file_id = item['id']
        file_name = item['name']

        # Access the file content
        request = drive_service.files().get_media(fileId=file_id)
        content = request.execute().decode('utf-8')
        with open('/content/Data/temp.txt', 'w', encoding='utf-8', errors='ignore') as file:
          file.write(content)
        directory = r'/content/Data'

        with ThreadPoolExecutor(max_workers=15) as executor:
            futures = [executor.submit(process_file, directory, filename) for filename in os.listdir(directory) if filename.endswith(".txt")]
            for future in futures:
                future.result()


        file_path = '/content/Data/temp.txt'

        # Read the file with UTF-8 encoding
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        Company_name = extract_company_name(content)
        print(Company_name)

        # Replace 'YOUR_YEAR' and 'YOUR_QUARTER' with the actual year and quarter you want
        year = '2022'
        quarter = 'Q3'
        content = extract_matched_texts(content)[0]

        extracted_info = extract_information(content)
        print(extracted_info)
        df = df.append({
            'File Name': file_name,
            'Company Name': Company_name,
            'Year': year,
            'Quarter': quarter,
            'Executive Details': extracted_info,
        }, ignore_index=True)
        count += 1
    #     break
    # break
    page_token = results.get('nextPageToken')
    if not page_token:
        break

df.to_csv('/content/Q3_2022.csv', index=False)

file_name = '/content/Q3_2022.csv'
files.download(file_name)
```

```python

for index, value in enumerate(df['Executive Details']):
    note_index = value.find('Note')
    explanation_index = value.find('Explanation:')
    expected_output_index = value.find('Expected output')

    remove_index = min(note_index if note_index != -1 else len(value),
                      explanation_index if explanation_index != -1 else len(value),
                      expected_output_index if expected_output_index != -1 else len(value))

    # Remove everything after the earliest occurrence
    result = value[:remove_index].rstrip()

    # Update the DataFrame column with the modified value
    df.at[index, 'Executive Details'] = result

```

**This is the code to get columns out of dicts**

```python

def process_detail(detail):
    if detail == 'No data found':
        return [{'Name': None, 'Age': None, 'Position held': None}]
    else:
        try:
            details = ast.literal_eval(detail)
            return details
        except (SyntaxError, ValueError):
            return [{'Name': None, 'Age': None, 'Position held': None}]

# Applying the function to create separate rows
new_rows = []
for index, row in df.iterrows():
    details = process_detail(row['Executive Details'])
    for detail in details:
        new_row = {
            'Company Name': row['Company Name'],
            'Year': row['Year'],
            'Quarter': row['Quarter'],
            'File Name': row['File Name'],
            'Name': detail.get('Name'),
            'Age': detail.get('Age'),
            'Position held': detail.get('Position held')
        }
        new_rows.append(new_row)

# Creating the new DataFrame
new_df = pd.DataFrame(new_rows)
new_df.to_csv('/content/2022_all_Q_final.csv', index=False)
from google.colab import files
file_name = '/content/2022_all_Q_final.csv'
files.download(file_name)
```
